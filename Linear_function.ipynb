```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

def plot_linear_function(slope, intercept):
    """
    English translation:

    Graphs a linear function with the given slope and intercept.

    Args:
    slope (float) The slope of the linear function.
    intercept (float) The intercept of the linear function.


    Spanish translation:
    Grafica una funcin lineal con la pendiente e interseccin dadas.

    Argumentos:
    pendiente (flotante) La pendiente de la funcin lineal.
    interseccin (floatante) La interseccin de la funcin lineal.
    """

    x = np.linspace(-10, 10, 400)  # Generates x values for the graph / Genera valores de x para el grfico
    y = slope * x + intercept  # Calculate the corresponding y values / Calcula los valores de y correspondientes

    plt.figure(figsize=(8, 6))  # Create a new shape for the chart / Crea una nueva figura para el grfico
    plt.plot(x, y, label=f'y = {slope}x + {intercept}')  # Graph the linear function / Grafica la funcin lineal
    plt.xlabel('x')  # Label the x-axis / Etiqueta el eje x
    plt.ylabel('y')  # Label the y-axis / Etiqueta el eje y
    plt.title('Linear function')  # Establece el ttulo del grfico
    plt.grid(True)  # Displays the grid on the chart / Muestra la cuadrcula en el grfico
    plt.axhline(0, color='black', linewidth=0.5)  # Draw the x-axis / Dibuja el eje x
    plt.axvline(0, color='black', linewidth=0.5)  # Draw the y-axis / Dibuja el eje y
    plt.legend()  # Displays the chart legend / Muestra la leyenda del grfico
    plt.show()  # Show the graph / Muestra el grfico

# Crea una interfaz interactiva para ajustar la pendiente y la interseccin
interact(plot_linear_function, slope=(-10.0, 10.0, 0.5), intercept=(-10.0, 10.0, 0.5))
```
