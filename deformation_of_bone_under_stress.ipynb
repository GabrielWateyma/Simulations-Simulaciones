```python
import numpy as np
import matplotlib.pyplot as plt

def bone_tolerance(forces):
    """
    Simulates bone deformation under tension.

    Args:
        forces: An array of applied forces.

    Returns:
        An array of corresponding deformations.
    """

    deformations = []
    elastic_limit = 200  # Example elastic limit force
    fracture_limit = 300  # Example fracture limit force

    for force in forces:
        if force <= elastic_limit:
            deformation = force * 0.01  # Proportional deformation
        elif force <= fracture_limit:
            deformation = elastic_limit * 0.01 + (force - elastic_limit) * 0.1 # Plastic deformation
        else:
            deformation = float('inf')  # Fracture

        deformations.append(deformation)

    return np.array(deformations)

# Generate force data
forces = np.linspace(0, 350, 100)

# Calculate deformations
deformations = bone_tolerance(forces)

# Plot
plt.plot(forces, deformations)
plt.xlabel("Force (N)")
plt.ylabel("Deformation (mm)")
plt.title("Bone Tolerance to Tension")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tolerancia_hueso(fuerzas):
    """
    Simula la deformación de un hueso bajo tensión.

    Args:
        fuerzas: Un array de fuerzas aplicadas.

    Returns:
        Un array de deformaciones correspondientes.
    """

    deformaciones = []
    limite_elastico = 200  # Fuerza límite elástica (ejemplo)
    limite_fractura = 300  # Fuerza límite de fractura (ejemplo)

    for fuerza in fuerzas:
        if fuerza <= limite_elastico:
            deformacion = fuerza * 0.01  # Deformación proporcional
        elif fuerza <= limite_fractura:
            deformacion = limite_elastico * 0.01 + (fuerza - limite_elastico) * 0.1 # Deformación plástica
        else:
            deformacion = float('inf')  # Fractura

        deformaciones.append(deformacion)

    return np.array(deformaciones)

# Generar datos de fuerza
fuerzas = np.linspace(0, 350, 100)

# Calcular deformaciones
deformaciones = tolerancia_hueso(fuerzas)

# Graficar
plt.plot(fuerzas, deformaciones)
plt.xlabel("Fuerza (N)")
plt.ylabel("Deformación (mm)")
plt.title("Tolerancia de un hueso a la tensión")
plt.grid(True)
plt.show()
```
